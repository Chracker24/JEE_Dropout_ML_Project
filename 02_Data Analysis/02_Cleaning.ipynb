# Cleaning
### to clean the original data for NaNs, proposed mapping and other substituitions


```python
import pandas as pd
import numpy as np
```

```python
jee = pd.read_csv("../Data/JEE_Dropout_After_Class_12.csv",delimiter=',')
jee.head()
```

```python
jee["coaching_institute"].fillna("Self-Study", inplace=True)
jee[["coaching_institute"]].sample(7)
```

```python
mapping_dict = {
  'school_board' : {
    'State' : 0,
    'CBSE' : 1,
    'ICSE' : 2
  },
  "coaching_institute" : {
    "FIITJEE" : 0,
    "Allen" : 0,
    "Local" : 1,
    "Self-Study":2
  },
  "family_income" : {
    "Low" : 0,
    "Mid" : 1,
    "High" : 2
  },
  "parent_education" : {
    "Upto 10th" : 0,
    "12th" : 1,
    "Graduate" : 2,
    "PG" : 3
  },
  "location_type" : {
    "Rural" : 0,
    "Semi-Urban" : 1,
    "Urban" : 2
  },
  "peer_pressure_level" : {
    "Low" : 0,
    "Medium" : 1,
    "High" : 2
  },
  "mental_health_issues" : {
    "No" : 0,
    "Yes" : 1
  },
  "admission_taken" : {
    "No" : 0,
    "Yes" : 1
  }
}
```

```python
for col, map_dict in mapping_dict.items():
    if col in jee.columns:
        jee[col] = jee[col].map(map_dict)
  
jee.sample(6)
```

```python
jee.isna().sum()
```

```python
jee.to_csv("../Data/JEE_Dropout_Cleaned.csv", index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
jee = pd.read_csv("01_Data/03_final/JEE_Dropout_Final.csv", delimiter=',')
sns.heatmap(pd.crosstab(jee["parental_support"],jee["dropout"]),annot=True)
plt.show()
```

```python
jee["parental_support"] = round(jee["parental_support"],1)
jee[["parental_support"]]
```

```python
plt.figure(figsize=(100,100))
sns.countplot(data=jee, x='parental_support', hue='dropout', palette='Set2')
plt.title("Dropout vs Parental Support")
plt.xlabel("Parental Support")
plt.ylabel("Number of Students")
plt.legend(title="Dropout", labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```
